#### Vector dot product 

$\alpha = \text{a}.\text{b} = \text{a}^{T}\text{b} = \sum_{i = 1}^n a_ib_i$

Dot product is also represented commonly as transpose of vector $\text{a}$ multiplied with $\text{b}$ 

```python
# Multiple ways to compute the dot product 
v1 = np.array([1, 2, 3, 5])
v2 = np.array([0, -2, 3, 5])

# First method 
dot_p1 = sum(np.multiply(v1, v2)) 

# Second method 
dot_p2 = np.dot(v1, v2)

# Third method 
dot_p3 = np.matmul(v1, v2) 

# Fourth method is to use for loop
dot_p4 = 0 

for i in range(0, len(v1)):
    dot_p4 = dot_p4 + v1[i] * v2[i] 
    
dot_p1, dot_p2, dot_p3, dot_p4
```

#### Properties of vectors (under dot product operation): 
a) Distributive: $a(b + c) = ab + ac$ and $a^T(b + c) = a^Tb + a^Tc$ \
b) Associatve: $a(b c) \ne(a \dot b)c$ and $a^T(b^Tc) \ne (a^Tb)^Tc$ \
c) Commutative: $ab = ba$

```python
## Distributive
n = 10 
a = np.random.randn(n) 
b = np.random.randn(n) 
c = np.random.randn(n)

# Dot product
res1 = np.dot(a, (b + c)) 
res2 = np.dot(a, b) + np.dot(a, c) 

# Verify
res1, res2
```

```python
## Associative 
num = 5
p = np.random.randn(num)
q = np.random.randn(num)
r = np.random.randn(num)

# Dot product 
r1 = np.dot(a, np.dot(b, c)) 
r2 = np.dot(np.dot(a, b), c)

# verify 
r1, r2 
```

```python
## Commutative 
num = 100
p = np.random.randn(num)
q = np.random.randn(num)  

# Dot product 
r1 = np.dot(p, q) 
r2 = np.dot(q, p) 

# Verify 
r1, r2 
```

#### Vector Length (geometric interpretation) 

$||v||^2 = (v_1)^2 + (v_2)^2$

#### Vector dot product: Geometrically 

$ \alpha = |a||b|\cos(\theta_{ab}) $ \
$ \alpha / |a||b| = \cos(\theta_{ab}) $ \
$ \theta_{ab} = \cos^{-1}(\alpha / |a||b|) $

```python
v1 = np.array([2, 4, -3])
v2 = np.array([0, -3, -3]) 

# Using the eqution we got to 
angle = np.arccos(np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2)))  # angle in rad

# Draw vectors
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot([0, v1[0]], [0, v1[1]], [0, v1[2]], 'b')
ax.plot([0, v2[0]], [0, v1[2]], [0, v1[2]], 'r') 

plt.axis((-6, 6, -6, 6)) 
plt.show()
```

#### Cauchy-Schwarz Inequality 

$|a^Tb| \le || {a} || \text{   } ||b||$ \
$||a|| \text{   } ||b|| \text{    } |\text{cos}(\theta_{ab})| \le ||a|| \text{   } ||b||$ 


```python
import numpy as np
 
a = np.random.randn(5) 
b = np.random.randn(5) 
c = np.random.randn(1) * a # dependent on a

# Compute the dot product 
aTb = np.dot(a, b)
aTc = np.dot(a, c) 

# Inequal since linearly independent 
print(f'{np.abs(aTb): .3f}, {np.linalg.norm(a) * np.linalg.norm(b): .3f}')

# Equal since linearly dependent
print(f'{np.abs(aTc): .3f}, {np.linalg.norm(a) * np.linalg.norm(c): .3f}') 
```

#### Vector Hadamard multiplication

Basically element wise multiplication between two vectors of same size. The output of Hadamard multiplication is a vector itself. 

```python
n = 10
v1 = np.random.randn(n)
v2 = np.random.randn(n) 
res = []

for i in range(0, len(v1)):
    res.append(v1[i] * v2[i])
    
print(res) 

# Direct way to do Hadamard multiplication usnig numpy 
res_2 = np.multiply(v1, v2) 

print(res_2) 
```

#### Vector product notations

Dot product produces a singular value or a 1 x 1 matrix. 

Outer produce produces a matrix of some size. 

Usual notations are: \
Dot product: $V^TW = 1 \times 1$ \
Outer product: $VW^T = N \times M$  

(V is of (N x 1) and W is of (M x 1) dimension)

```python
v1 = [1,2,3]
v2 = [4,5,6]

# Direct way to find the output using numpy
print(np.outer(v1, v2)) 

# Using handcoded for loops
result_mat = np.zeros((len(v1), len(v2)))
for i in range(0, len(v1)):
    for j in range(0, len(v2)):
        result_mat[i, j] = v1[i] * v2[j]
        
print(result_mat)
```

#### Cross Product 

```python
# Create vectors 
v1 = [-3, 2, 5]
v2 = [4, -3, 0] 

# direct method using numpy
v3a = np.cross(v1, v2)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Drawing a plane defined by span of v1 and v2
xx, yy = np.meshgrid(np.linspace(-10, 10, 10), np.linspace(-10,10,10))
z1 = (-v3a[0]*xx - v3a[1]*yy) / v3a[2]
ax.plot_surface(xx, yy, z1) 

ax.plot([0, v1[0]], [0, v1[1]], [0, v1[2]], 'k')
ax.plot([0, v2[0]], [0, v2[1]], [0, v2[2]], 'k') 
ax.plot([0, v3a[0]], [0, v3a[1]], [0, v3a[2]], 'r') 

ax.view_init(azim=150, elev=45)
plt.show()
```

#### Hermitian transpose (conjugate transpose)

```python
z = complex(3,4)

print(np.linalg.norm(z))

# Transpose
print(np.transpose(z) * z) 

# Apply Hermitian transpose
print(np.transpose(z.conjugate())* z) 

# Complex vector
v = np.array([3, 4j, 5 + 2j, complex(2, -5)]) 
print(v.T)
print(np.transpose(v))
print(np.transpose(v.conjugate())) 
```

### Concept of subspaces

Subspace have algebraic and geometric interpretations:

Algebraic: 
Consider $\lambda \in \mathbb{R}$ a scalar and $v$ is a vector, then any $\lambda v$ and $v$ will reside in the same vector space.

This concept can be further extended as a linear combination of form $\lambda v$ + $\beta w$, that is creating a subspace consisting of linear combinations.

Properties of subspaces: \
a) Should be closed under addition and scalar multiplication \
b) Contain the zero vector.

Mathematically: 
$\forall v, w \in V; \forall \lambda, \beta \in \mathbb{R}; \lambda v + \beta w \in W $

#### Span of vectors

$span(\{v_1, \dots, v_n\}) = \alpha_1v_1 + \dots + \alpha_nv_n, \alpha \in \mathbb{R}$


Span of a set of vectors is all possible linear combinations of all the vectors present in the set. This spans streches to infinity. 

```python
# Set S
S1 = np.array([1,1,0])
S2 = np.array([1,7,0])

# vectors v and w
v = np.array([1,2,0])
w = np.array([3,2,1]) 

# vectors
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot([0, S1[0]], [0, S1[1]], [.1, S1[2] + .1], 'r', linewidth=3)
ax.plot([0, S2[0]], [0, S2[1]], [.1, S2[2] + .1], 'black', linewidth=3)

ax.plot([0, v[0]], [0, v[1]], [.1, v[2] + .1], 'g', linewidth=3)
ax.plot([0, w[0]], [0, w[1]], [0, w[2]], 'b') 

xx, yy = np.meshgrid(range(-15, 16), range(-15, 16)) 
cp = np.cross(S1, S2) 
z1 = (-cp[0]*xx - cp[1]*yy)*1/cp[2]
ax.plot_surface(xx, yy, z1) 

plt.show()
```
